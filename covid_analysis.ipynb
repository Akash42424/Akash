```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("owid-covid-data.csv")
df.head()
```

```python
df["month"]=pd.to_datetime(df["date"]).dt.month
df["year"]=pd.to_datetime(df["date"]).dt.year
df2=df[df.location=="India"]
df2
```

```python
df2.reset_index(drop=True,inplace=True)
df2.tail()
```

```python
df3=df2[(df2.month>3) & (df2.year!=2019)]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5),dpi=150)
plt.bar(df3.date,df3.total_cases)
plt.grid()
plt.xticks(rotation="vertical")
plt.xlabel("date")
plt.ylabel("total_cases")
plt.title("covid in India")
plt.show()
```

```python

plt.figure(figsize=(10,5),dpi=150)
plt.bar(df3.date,df3.new_cases)
plt.grid()
plt.xticks(rotation="vertical")
plt.xlabel("date")
plt.ylabel("new_cases")
plt.title("covid in India")
plt.show()
```

```python
plt.figure(figsize=(10,5),dpi=150)
plt.bar(df3.date,df3.total_deaths)
plt.xticks(rotation="vertical")
plt.xlabel("date")
plt.ylabel("total_deaths")
plt.show()
```
