### comparing between Countries  (Usa ,uk,russia,china) in last 2 decade

```python
country=["United States","Russia","China","Germany","Australia"]
df_comp=pd.DataFrame({})
for c in country:
    x=df2[(df2["Year"]>1995) &(df2["Team"]==c) &(df2["Season"]=="Summer")]
    df_comp=pd.concat([df_comp,x])
```

```python
df_comp
```

```python
sns.countplot(df_comp["Team"],hue=df_comp["Sex"])
plt.title("Medal winner in last Two decade",fontsize=15,color="red")
plt.show()
```

```python
df_comp_gold=df_comp[df_comp["Medal"]=="Gold"]
```

```python
sns.countplot(df_comp_gold["Team"],hue=df_comp_gold["Sex"])
plt.title("Gold winner in last Two decade",fontsize=15,color="red")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df_comp["Team"],hue=df_comp["Year"])
plt.title("Medal winner in last Two decade",fontsize=15,color="red")
plt.show()
```
